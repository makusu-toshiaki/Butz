```python
import requests
import records
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import sqlite3
```

```python
db = records.Database('sqlite:///ichirann.db')

db.query("""CREATE TABLE IF NOT EXISTS Countries (
         country_name text PRIMARY KEY,
         content text)""")

def store_sekai(country_name, content):
    try:
        db.query("""INSERT INTO Countries (country_name, content)
                    VALUES (:country_name, :content)""", country_name=country_name, content=content)
    except IntegrityError as ie:
        # この国名は存在するので何もしない
        pass


url = 'https://ja.wikipedia.org'
r = requests.get(url+'/wiki/国の一覧')

html_soup = BeautifulSoup(r.text, 'html.parser')
# print(html_soup.text)
table = html_soup.find('table',class_= 'wikitable')
td = table.find_all('td')
# print(td)
for country_name in td:
    if '語' in country_name.text:
        continue
    print(country_name.text)
    content = country_name.find_all('a')
    for href in content:
        if '.svg' in href:
            continue
        new_url = href.get('href')
    r2 = requests.get(url+new_url)
    print(r2)
    store_sekai(country_name.text, r2.text)
```
