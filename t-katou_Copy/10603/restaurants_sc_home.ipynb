```python
conn = None

# データベースに接続する
def connect():
    global conn
    conn = sqlite3.connect("./Restaurants_Reviews_home.db")
    
# コネクションを断つ
def close():
    conn.close()
    
# テーブル作成
def create_table_Restaurants():
    # DROP=消す.
    conn.execute("DROP TABLE IF EXISTS Restaurants")
    conn.execute("""CREATE TABLE Restaurants (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            rest_name TEXT
    )""")
    
def create_table_Reviews():
    # DROP=消す.
    conn.execute("DROP TABLE IF EXISTS Reviews")
    conn.execute("""CREATE TABLE Reviews (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            rest_id INT,
            review TEXT,
            rate FLOAT
    )""")
    
# loadは積むという意味
def load_Restaurants(rest_name):
    conn.execute("INSERT INTO Restaurants (rest_name) VALUES (?)", (rest_name,))
    conn.commit()
    
def load_Reviews(rest_id, review, rate):
    conn.execute(
        "INSERT INTO Reviews (rest_id,review,rate) VALUES (?,?,?)", (rest_id,review,rate))
    conn.commit()
```

```python
def first(new_url):
    try:
         # 件数クリックしたurl
        burger_r = requests.get(new_url)
        burger_soup = BeautifulSoup(burger_r.text, 'html.parser')
#         print(burger_soup)
        print("first")
        return burger_soup
    except:
        time.sleep(5)
        first(new_url)
        
def second(new_url,j):
    try:
        # j+1ページ目のurl
        review_r = requests.get(new_url+"COND-0/smp1/?smp=1&lc=0&rvw_part=all&PG="+str(j+1))
        review_soup = BeautifulSoup(review_r.text, 'html.parser').find_all("div", class_ ="rvw-item js-rvw-item-clickable-area")
#         print(review_soup)
        return review_soup
    except:
        time.sleep(5)
        second(new_url,j)
        
def get_review_rate(review):
    try:
        review_last = requests.get(review)
        review_last_soup = BeautifulSoup(review_last.text, 'html.parser')
#         print("third")
        rest_review = review_last_soup.find("div", class_ ="rvw-item__rvw-comment").text
        if review_last_soup.find("b", class_ ="c-rating__val c-rating__val--strong").text == "-":
            return [rest_review,"0"]
        rate = float(review_last_soup.find("b", class_ ="c-rating__val c-rating__val--strong").text)

        return [rest_review,rate]
    except:
        print("失敗")
        time.sleep(5)
        get_review_rate(review)
```
