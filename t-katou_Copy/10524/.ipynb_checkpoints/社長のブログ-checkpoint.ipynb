```python
from bs4 import BeautifulSoup
import requests
import re
import sqlite3
```

```python
url = "https://www3.ttc-net.co.jp/member-page/wp-login.php?redirect_to=/"
r = requests.get(url)
bs = BeautifulSoup(r.text,"html.parser")

# print(bs.text)
my_seassion = requests.Session()
# print(my_seassion)
r = my_seassion.get(url)
r = my_seassion.post(url,data = {"log": "t-katou", "pwd": "1dksa09181220"})

print(r.text)
```

```python
ttc = BeautifulSoup(r.text,"html.parser")
# print(ttc)
li = ttc.find("li", class_="menu-item menu-item-type-custom menu-item-object-custom menu-item-25613")
# print(li)
li2 = li.find("a")
# print(li2)
a = li2.get("href")
print(a)
# r_syatyo = requests.get(a)
r_syatyo = my_seassion.get(a)
bs_syatyo = BeautifulSoup(r_syatyo.text,"html.parser")

```

```python
# print(bs_syatyo)
# print(bs_syatyo.find_all("h2"))
def syatyo_blog(bs_syatyo):
    new_url = []
    for li_syatyo in bs_syatyo.find_all("h2"):
        a_syatyo = li_syatyo.find("a")
#         print(a_syatyo)
        url_syatyo = a_syatyo.get("href")
        new_url.append(url_syatyo)
    return new_url

new_url = syatyo_blog(bs_syatyo)
print(new_url)
```

```python
dic = {}
for blog_url in new_url:
    blog_page = my_seassion.get(blog_url)
    blog_text = BeautifulSoup(blog_page.text,"html.parser")
#     print(blog_text)
    blog_title = blog_text.find("h1",class_="entry-title")
    blog_title2 = re.sub("\u3000","",blog_title.text)
#     print(blog_title.text)
    blog_content = blog_text.find("div",class_="entry-content")
    blog_content2 = re.sub("\u3000","",blog_content.text)
#     print(blog_content.text)
    dic[blog_title2] = blog_content2
```

```python
print(dic)
```
